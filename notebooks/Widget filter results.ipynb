```python
# collect all the possible categories and build a section widget
import ipywidgets
from IPython.display import display,clear_output
#url = '%s/categorize/index.csv'%server
#df = pd.read_csv(url)
categ = all_coords['Institution'].unique().tolist()
# build the widget
resp = ipywidgets.Dropdown(options=categ,description='Institution:',rows=10,columns=2)
display(resp)
def on_value_change(change):
    global term
    global resp
    global cat
    cat=change['new']
    #global change
    #url_var='%s/categorize/%s/index.csv'%(server,change['new'])
    #df_var=pd.read_csv(url_var)
    #term=df_var['Category'].values
    term = all_coords.loc[all_coords['Institution']==change['new'],['Dataset ID']]['Dataset ID'].unique().tolist()
    term = ipywidgets.SelectMultiple(options=term,description='%s dataset:'%(change['new']),rows=20,columns=20,layout=ipywidgets.Layout(width='100%', height='200px'))
    clear_output(wait=True)
    display(resp,term)
    #print(url_var)
resp.observe(on_value_change, names='value')
```
